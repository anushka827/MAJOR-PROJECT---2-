```python
# embeddings.py logic in Colab
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import Chroma

# Replace with your OpenAI API key
OPENAI_API_KEY = "YOUR_API_KEY"

# Initialize embeddings
embeddings = OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY)

# Create or load vectorstore
vectorstore = Chroma(persist_directory=chroma_db_path, embedding_function=embeddings)
print("Vectorstore initialized ✅")
```

```python
# ===========================
# STEP 3: Build RAG Engine
# ===========================
from langchain.chains import RetrievalQA
from langchain.llms import OpenAI

# Initialize LLM
llm = OpenAI(temperature=0.0, openai_api_key=OPENAI_API_KEY)

# Create retriever from vectorstore
retriever = vectorstore.as_retriever()

# Create RetrievalQA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever
)

# Function to ask questions
def ask_question(question):
    return qa_chain.run(question)

print("RAG Engine ready ✅")
```

```python
import streamlit as st
import threading
import os

st.title("AI RAG Chatbot (Colab Demo)")

# ---------------------------
# Function to ask question
# ---------------------------
def ask_question(user_input):
    from langchain.embeddings import OpenAIEmbeddings
    from langchain.vectorstores import Chroma
    from langchain.chains import RetrievalQA
    from langchain.llms import OpenAI

    OPENAI_API_KEY = "YOUR_API_KEY"
    chroma_db_path = "/content/mydrive/MyDrive/ai-rag-chatbot/data/chroma_db"

    # Load embeddings and vectorstore on-demand
    embeddings = OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY)
    vectorstore = Chroma(persist_directory=chroma_db_path, embedding_function=embeddings)

    llm = OpenAI(temperature=0, openai_api_key=OPENAI_API_KEY)
    retriever = vectorstore.as_retriever()
    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=retriever
    )

    return qa_chain.run(user_input)

# ---------------------------
# Input box
# ---------------------------
user_input = st.text_input("Ask me anything:")

if user_input:
    with st.spinner("Thinking... ⏳"):
        answer = ask_question(user_input)
    st.write("Answer:", answer)
```
